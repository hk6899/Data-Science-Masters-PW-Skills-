We can take the average score of the individual scores and visualise with the other columns and understand the relationship

```python
df_stud['total_score']=(df_stud['math_score']+df_stud['reading_score']+df_stud['writing_score'])
df_stud['average']=df_stud['total_score']/3
df_stud.head()
```

```python
## Explore More Visualization
fig,axis=plt.subplots(2,2,figsize=(15,7))
plt.subplot(321)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,color='g')
plt.subplot(322)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,hue='gender')
plt.subplot(323)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,hue='lunch')
plt.subplot(324)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,hue='test_preparation_course')
plt.subplot(325)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,hue='parental_level_of_education')
plt.subplot(326)
sns.histplot(data=df_stud,x='average',bins=30,kde=True,hue='race_ethnicity')


plt.show()
```

The above figure shows how the various features affect the perfomance of the students.Studens who take a standard lunch and who does the test preparation course are scoring well. Girls score better than boys. Parents who have masters degree, their children score well
