 What is the significance level commonly used in hypothesis testing?

The significance level commonly used in hypothesis testing is 0.05, which is also known as the alpha level. This means that if the p-value (probability value) of a statistical test is less than 0.05, the null hypothesis is rejected, and the alternative hypothesis is accepted as being statistically significant.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import scipy.stats as stats

# Define the samples (replace with your own data)
sample1 = [5, 7, 6, 8, 9]
sample2 = [4, 6, 7, 5, 8]

# Set the significance level (alpha)
alpha = 0.05

# Perform the hypothesis test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Print the results
print("Hypothesis Test Results:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Q8. Define p-value and explain its significance in hypothesis testing.

p value is used to dsecribe how likely you are found to have a particular set of observations if the null hypothesis were true. In simpler terms, the p-value represents the likelihood of observing the data if the null hypothesis is correct.It's important to note that the p-value is not a measure of the magnitude or practical significance of an effect. Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define the degrees of freedom
df = 10

# Generate x-values for the plot
x = np.linspace(-4, 4, 500)

# Calculate the probability density function (PDF) for the Student's t-distribution
pdf = t.pdf(x, df)

# Create the plot
plt.plot(x, pdf, label='t-distribution (df = 10)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title("Student's t-distribution (df = 10)")
plt.legend()

# Show the plot
plt.show()
```

Q12: What is t-statistic? State the formula for t-statistic.

A t-statistic, also known as a t-value or t-score, is a statistical measure that is commonly used in hypothesis testing. It quantifies the difference between a sample mean and a population mean, taking into account the variability in the sample data
One-sample t-test: In this scenario, the null hypothesis assumes that the sample mean is equal to a hypothesized population mean. The t-statistic is calculated as:

t = (sample_mean - hypothesized_mean) / (sample_standard_deviation / sqrt(sample_size))

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

To test the hypothesis, we can use a one-sample t-test. Let's define the null and alternative hypotheses:

Null hypothesis (H0): The new drug does not decrease blood pressure by 10 mmHg.
Alternative hypothesis (H1): The new drug decreases blood pressure by 10 mmHg.

We will use a significance level () of 0.05.

To conduct the t-test, we need to calculate the t-statistic using the following formula:

t = (sample mean - hypothesized mean) / (standard deviation / sqrt(sample size))

In this case:
Sample mean (x) = 8 mmHg
Hypothesized mean () = 10 mmHg
Standard deviation () = 3 mmHg
Sample size (n) = 100

Let's calculate the t-statistic:

t = (8 - 10) / (3 / sqrt(100))
t = -2 / (3 / 10)
t = -20 / 3
t  -6.67

Now, we need to determine the critical t-value based on the significance level and degrees of freedom (n - 1). Since the sample size is 100, the degrees of freedom is 99.

Using a t-table or statistical software, we find that the critical t-value for a significance level of 0.05 and 99 degrees of freedom is approximately 1.984.

Since the calculated t-value (-6.67) is smaller (in magnitude) than the critical t-value (-1.984), we can reject the null hypothesis.

Therefore, based on the data from the clinical trial, there is evidence to suggest that the new drug decreases blood pressure by more than 10 mmHg.
